```python
'''
This function filters all images.

Routes should be in absolute path format.

params
------
ruta_imagenes: Folder 
ruta_f1: Stores images that were discarted by the first filter in this folder.
ruta_f2: Stores images that were discarted by the second filter in this folder.
ruta_f3: Stores images that were discarted by the third filter in this folder.
ruta_r: Stores images that passed all the filters.
count_max: Max amount of images that will be stored in the folder. 

return
------
count_pass: the number of 
'''
def filtros_concatenados(ruta_imagenes, ruta_f1, ruta_f2, ruta_f3, ruta_r, count_max):
  
  # numero de imagenes que pasaron
  count_pass = 0
    
  mi_ruta = ruta_imagenes  
  for count,a in enumerate(os.listdir(mi_ruta)):
        
    # evalua si salimos del bucle o no
    if count_max <= count_pass:
      break

    try:
          
      # filter 1 (Detectron): At least one should be a car, bus, or truck 
      im = cv2.imread(mi_ruta +"/"+a)
      outputs = predictor(im)
      aaa = outputs["instances"].pred_classes
      if (2 in aaa or 5 in aaa or 7 in aaa):
          
        # filter 2 (Detectron): The biggest object must be one of the following: car, bus, or truck 
        objects_area = outputs["instances"].pred_boxes.area()
        objects_coor = outputs["instances"].pred_boxes.tensor
        objects_area_numpy = objects_area.cpu().numpy()
        objects_coor_numpy = objects_coor.cpu().numpy()
        max_index = np.argmax(objects_area_numpy)
        if 2 == int(aaa[max_index]) or 5 == int(aaa[max_index]) or 7 == int(aaa[max_index]):        
          
          # filter 4 (OpenCV): cut image
          eee = objects_coor_numpy.tolist()[max_index]
          im = im[int(eee[1]):int(eee[3]), int(eee[0]):int(eee[2])]
          
          # filtro 3 (Mobilenet): Only pass images of cars taker from the outside
          b = cv2.resize(im,(160,160))
          img_array = np.expand_dims(b, axis=0)
          # 0: interior
          # 1: exterior
          # 2: llanta
          predictions = model.predict(img_array,verbose=0)
          score = tf.nn.softmax(predictions[0])
          if (int(np.argmax(score)) == 1):
                  
            os.chdir(ruta_r)
            cv2.imwrite(a, im)
     
            # paso una imagen mas por todos los filtros. Aumenta el contador
            count_pass = count_pass + 1
     
      
    except Exception as e: 
      print(e)
    
  return count_pass
```
